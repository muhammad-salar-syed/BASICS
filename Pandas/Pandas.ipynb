```python
def grouper(df,idx,col):
    if 80<=df[col].loc[idx]<=90:
        return '80-90'
    elif 50<=df[col].loc[idx]<=60:
        return '50-60'
    else:
        return 'others'
```

## Merge

```python
import pandas as pd
df1=pd.DataFrame({
    'city':['new york','chicago','orlando'],
    'temperature':[21,14,35]
})
df1
```

```python
df2=pd.DataFrame({
    'city':['chicago','new york','orlando'],
    'humidity':[65,68,75]
})
df2
```

```python
df=pd.merge(df1,df2,on='city')
df
```

```python
# join
df1=pd.DataFrame({
    'city':['new york','chicago','orlando','baltimore'],
    'temperature':[21,14,35,38]
})
df1
```

```python
df2=pd.DataFrame({
    'city':['chicage','new york','san diego'],
    'humidity':[65,68,71]
})
df2
```

```python
df3=pd.merge(df1,df2,on='city',how='outer')
df3
```

```python
df4=pd.merge(df1,df2,on='city',how='inner')
df4
```

```python
df5=pd.merge(df1,df2,on='city',how='right')
df5
```

```python
df6=pd.merge(df1,df2,on='city',how='left')
df6
```

```python
# indicator
df7=pd.merge(df1,df2,on='city',how='outer',indicator=True)
df7
```

```python
# suffix
df8=pd.DataFrame({
    'city':["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df8
```

```python
df9=pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df9
```

```python
df10=pd.merge(df8,df9,on="city",how="outer", suffixes=('_first','_second'))
df10
```

```python
df11 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35]
})
df11.set_index('city',inplace=True)
df11
```

```python
df12 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75]
})
df12.set_index('city',inplace=True)
df12
```

```python
# join
df11.join(df12,lsuffix='_l', rsuffix='_r')
```

```python
# Using CustomBusinessDay to generate US holidays calendar frequency
from pandas.tseries.holiday import USFederalHolidayCalendar
from pandas.tseries.offsets import CustomBusinessDay

us_cal = CustomBusinessDay(calendar=USFederalHolidayCalendar())

rng1 = pd.date_range(start="7/1/2017",end="7/23/2017", freq=us_cal)
rng1
```

```python
# AbstractHolidayCalendar
from pandas.tseries.holiday import AbstractHolidayCalendar, nearest_workday, Holiday
class myCalendar(AbstractHolidayCalendar):
    rules = [
        Holiday('My Birth Day', month=4, day=15),#, observance=nearest_workday),
    ]
    
my_bday = CustomBusinessDay(calendar=myCalendar())
pd.date_range('4/1/2017','4/30/2017',freq=my_bday)
```

```python
egypt_weekdays = "Sun Mon Tue Wed Thu"
b = CustomBusinessDay(weekmask=egypt_weekdays)
pd.date_range(start="7/1/2023",periods=20,freq=b)
```
